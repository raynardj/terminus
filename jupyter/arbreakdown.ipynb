```python
# client id and client group mapping
gpnames="A,C"
gpdf=pd.read_excel('cplist.xlsx',parse_cols=gpnames).rename(columns={u"客户编号":u"客户"}).set_index(u"客户")
gpdf
```

```python
# join ar dataframe with group data frame(providing it's already index)
df=ardf.join(gpdf,on=u"客户")
df
```

```python
# check the join result
df[df[u"是否CP"]==u"国网项目"]
```

```python
# Try one pivot table 
pd.pivot_table(df,index=u"BU",columns=u"City",values=u"本币金额").fillna(value=0)
```

```python
import numpy as np
```

```python
# Cut to slots by "bins" bondary
bins=[-600,90,180,200000]
cates=[u"0~90days",u"90~180days",u"180days+"]
df["ASlots"]=pd.cut(df[df.columns[6]],bins,labels=cates)
# Refill the BU column with group tag
df[u"BU"][df[u"是否CP"].isnull()==False]=u"其他CP"
df[u"BU"][df[u"是否CP"]==u"国网项目"]=u"国网项目"
df[u"BU"][df[u"是否CP"]==u"移动项目"]=u"移动项目"
# Check the BU possible values
set(df[u"BU"])
```

### Create Pivot Table

```python
dfpv=df.drop(df.index[df[u"BU"] ==u"SBD"])
dfpv=dfpv.drop(df.index[df[u"BU"] ==u"Third Party"])
dfpv=dfpv.drop(df.index[df[u"BU"] ==u"Retail"])
pv=pd.pivot_table(dfpv,index=u"BU",columns=u"ASlots",values=u"本币金额").fillna(value=0)
pv
```
