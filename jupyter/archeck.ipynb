### The Dictionary: Translate the field names to eng names

```python
od_fields=dict({
u'城市':"city",
u'客户编号':"cid",
u'客户名称':"cnm",
u'客户类型':"ctyp",
u'逾期天数':"oday",
u'账期':"term",
u'未核销金额':"amt",
u'下单日期':"odt",
u'发货日期':"ddt",
u'原始销售订单':"oodr",
u'销售订单':"odr",
u'销售订单金额':"oamt",
u'交货单':"dn",
u'开票日期':"idt",
u'手工发票号':"inv",
u'发票金额':"iamt",
u'发票抬头':"itt",
u'已核销金额':"aamt",
u'发票类型':"ityp",
u'付款方式':"pay",
})
```

```python
oddf.columns
```

```python
oddf=oddf.rename(columns=od_fields)
oddf
```

```python
from datetime import date
```

```python
rkeys=["city","amt","oamt","damt","odt","ddt","idt","odr","oodr","inv","dn","itt","cid"]
for rowid in range(len(oddf)):
    rd=oddf.loc[rowid].to_dict()
    rd2=dict()
    for k in rkeys:
        if rd.has_key(k):
            if k in ["amt","oamt","damt"]:# Process Number, float, and round to 2 decimal points
                rd2.update({k:round(rd[k],2)})
            elif k in ["cid","odr","oodr","dn","inv"]: #Process Serial Numbers
                rd2.update({k:str(rd[k])})
            elif k in ["odt","ddt","idt"]:#Process date format
                print(type(rd[k]))
                rd2.update({k:rd[k].to_datetime()})
            else:
                rd2.update({k:rd[k]})
    print rd2
```
