```python
from predef import *
from pymongo import ASCENDING
from pymongo import DESCENDING
import pickle
```

```python
db=get_db("foundation")
```

```python
age_dfs=list(db.armonth_age_rp.find({}).sort("month",DESCENDING).sort("year",DESCENDING).limit(10))
true_span=len(age_dfs)
alldf=pd.DataFrame()
for month in range(true_span):
    df=pickle.loads(age_dfs[month]["df"])
    alldf["year:%s,month:%s"%(age_dfs[month]["year"],age_dfs[month]["month"])]=df.sum(axis=0)
alldf
```

```python
alldf=alldf.drop("0~10 Days")
```
