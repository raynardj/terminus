导入数据 和 函数， 大概需要十几秒，只要导入一次即可

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import numpy as np

def similar(n,ad):
    srdf=ad
    rlist=list()
    for k in ad.index:
        try:
            rlist.append(fuzz.ratio(ad.loc[n][0],ad.loc[k][0]))
        except:
            rlist.append(0)
    srdf["match"]=rlist
    return srdf
def similar_w(w,ad):
    srdf=ad
    rlist=list()
    for k in ad.index:
        try:
            rlist.append(fuzz.ratio(w,ad.loc[k][0]))
        except:
            rlist.append(0)
    srdf["match"]=rlist
    return srdf
def sn(n):
    srdf=similar(n,ad).sort_values(by="match",ascending=0)
    return srdf
def sw(w):
    srdf=similar_w(w,ad).sort_values(by="match",ascending=0)
    return srdf
import pandas as pd
ex=pd.read_excel('addresswork.xlsx')
ex["_id"]=range(len(ex))
adddf=ex.replace({' ':''},regex=True)[u"销售订单收货地址"]
adddf_set=pd.DataFrame(list(set(adddf)),index=range(len(set(adddf))))
ad=adddf_set.fillna(0)
```
